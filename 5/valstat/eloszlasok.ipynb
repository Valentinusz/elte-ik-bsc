# Fontosabb valószínűségi eloszlások

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# locale for decimal comma
import locale
locale.setlocale(locale.LC_NUMERIC, "de_DE")

# enable TeX for matplotlib
plt.rcParams['text.usetex'] = True
plt.rcParams['axes.formatter.use_locale'] = True
```

## Diszkrét eloszlások

### Binomiális eloszlás

```python
fig, ax = plt.subplots(1, 1)
x = np.arange(0, 40)

binom1 = binom(20, 0.5)
binom2 = binom(20, 0.7)
binom3 = binom(40, 0.5)

ax.plot(x, binom1.pmf(x), 'bo', ms=5, label=rf'n=20, p=0,5')
ax.plot(x, binom2.pmf(x), 'go', ms=5, label=rf'n=20, p=0,7')
ax.plot(x, binom3.pmf(x), 'ro', ms=5, label=rf'n=40, p=0,5')
ax.legend()
```
